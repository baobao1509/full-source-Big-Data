```python
import os
import sys

# 1. CẤU HÌNH ĐƯỜNG DẪN
java_home = r'C:\Progra~1\Java\jdk-17' # Dùng bản 17 là chuẩn nhất
hadoop_home = r'C:\hadoop'

os.environ['JAVA_HOME'] = java_home
os.environ['HADOOP_HOME'] = hadoop_home
os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable

# Cập nhật PATH
os.environ['PATH'] = os.path.join(java_home, 'bin') + os.pathsep + \
                     os.path.join(hadoop_home, 'bin') + os.pathsep + \
                     os.environ['PATH']

from pyspark.sql import SparkSession

try:
    print("--- Đang khởi tạo Spark bản fix cho Python 3.12... ---")
    spark = SparkSession.builder \
        .master("local[*]") \
        .config("spark.driver.host", "127.0.0.1") \
        .config("spark.driver.bindAddress", "127.0.0.1") \
        .config("spark.driver.extraJavaOptions", "-Djava.net.preferIPv4Stack=true") \
        .getOrCreate()
    
    print("🚀 CUỐI CÙNG CŨNG XONG! Spark đã chạy.")
    print("Version:", spark.version)
    spark.range(3).show()
except Exception as e:
    print(f"❌ Vẫn lỗi: {e}")
    print("\n--- NẾU VẪN LỖI, HÃY CHẠY LỆNH NÀY TRONG TERMINAL ---")
    print("pip show pyspark")
```
